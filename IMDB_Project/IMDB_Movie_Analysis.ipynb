```python
import pandas as pd

df = pd.read_csv("movie_metadata.csv")

df.head()
```

```python
df.shape
df.columns
df.info()      ##datatype and missing value summary
df.describe()    #statistics of numeric columns
```

```python
df.isnull().sum()
```

```python
df= df.dropna(subset=["movie_title","imdb_score"])    #drop where movie title and imdb score is missing 
```

We’ll only fill important numeric columns that are useful for analysis with mean.
These have missing values but are valuable:
duration (movie length)
budget (movie cost)
gross (movie earnings)

df["duration"] → selects the column “duration”
.fillna(...) → fills missing (NaN) values with something
df["duration"].mean() → calculates the average duration
inplace=True → makes the change directly in the same table (no need to reassign)

```python
df["duration"].fillna(df["duration"].mean(),inplace =True)
df["budget"].fillna(df["budget"].mean(), inplace=True)
df["gross"].fillna(df["gross"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df["num_user_for_reviews"].fillna (df["num_user_for_reviews"].mean(), inplace = True)
df["aspect_ratio"].fillna(df["aspect_ratio"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

Cleaning done !! DATA IS READY NOW FOR DATA ANALYSIS :)

```python
#Top 10 Highest-Rated Movies --Which movies in the dataset have the highest IMDB ratings?

df[["movie_title", "imdb_score"]].sort_values(by ="imdb_score", ascending = False).head(10)
```

```python
#What year had the highest average IMDB ratings?-  which years had the best-rated movies on average. 
#This shows how movie quality (based on ratings) changed over time.

df.groupby("title_year")["imdb_score"].mean().sort_values(ascending = False).head(10)
```

```python
#1. Average IMDB Score per Genre (first genre only)
df["main_genre"]= df["genres"].str.split('|').str[0]

#avg rating per main genre
df.groupby("main_genre")["imdb_score"].mean().sort_values(ascending=False)
df["main_genre"].value_counts()
```

```python
df.groupby("main_genre")["imdb_score"].mean().sort_values(ascending=False)
```

```python
plt.savefig("genre_avg_ratings.png")
```

```python

import matplotlib.pyplot as plt

# Group and sort
avg_ratings = df.groupby("main_genre")["imdb_score"].mean().sort_values(ascending=True)

# Plot
plt.figure(figsize=(10, 8))
plt.barh(avg_ratings.index, avg_ratings.values, color='cornflowerblue')
plt.xlabel("Average IMDB Score")
plt.ylabel("Genre")
plt.title("Average IMDB Score by Movie Genre")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("genre_avg_ratings.png")  # Save it for GitHub
plt.show()
```

```python
#Profitability by Genre

df["profit"] = df["gross"]-df["budget"]
```

```python
genre_profit = df.groupby("main_genre")[["budget","gross","profit"]].mean().sort_values(by="profit", ascending=False)
genre_profit.head()
```

```python
plt.savefig("genre_profit.png")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(genre_profit.index, genre_profit["profit"], color='mediumseagreen')
plt.xlabel("Average Profit")
plt.title("Average Profit by Genre")
plt.grid(True)
plt.tight_layout()
plt.savefig("avg_profit_by_genre.png")
plt.show()
```
